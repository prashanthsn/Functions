6. Explain the concept of generators in Python and how they are defined.
- Generator functions in Python are a way to create iterators in a more concise and memory efficient way. It is defined by 'yield'  

```python
def a():
    yield 1
    yield 2
    yield 3
# Using the generator
generator = a()
print(next(generator))  # Output: 1
print(next(generator))  # Output: 2
print(next(generator))  # Output: 3
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map(): It applies a given function to all the items in an iterable and returns an iterable map object of result.

```python
a=[1,2,3,4,5]
result = map(lambda x: x**2, a)
print(list(result))
```

- reduce(): It is a process where we reduce a list of items to a single value(cumulative value)


```python
from functools import reduce
numbers = [1, 2, 3, 4, 5]
total_sum = reduce(lambda x, y: x + y, numbers)
print(total_sum)
```

- filter(): It filters elements containing only the elements that met the specified criteria.

```python
a=[1,2,3]
result = filter(lambda x: x%2==0, a) # filters only the number which is divisible by 2 and shows that element only
print(list(result))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime():
    return[num for num in range(1,201)if is_prime(num)]
primes=prime()
print(f"The prime numbers are {primes}")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
# Using the generator for Fibonacci sequence
fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i
for power in power_of_two(3):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,-  if the value of the
order is smaller than 100,00 .

Write a Python program using lambda and map.

```python
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python3, Bernd Klein", 3, 24.99]
]

# here, we calculate the total price and increase by 10 if it's less than 100
calculate_total = lambda order: (order[0], order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0))

# Use map to apply the lambda function to all orders
result = list(map(calculate_total, orders))
print(result)
```
